```python
# Importamos las librer√≠as necesarias
#Fernando
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np  # import numpy para el manejo de arrays
from sklearn.model_selection import train_test_split
```

```python
#Cargamos el Dataset
df = pd.read_csv('C:/Users/Fer/Documents/Fernando/archivospython/WA_Fn-UseC_-HR-Employee-Attrition.csv', sep=',')
df
```

```python
#Transofrmamos variables a binarias

#Transformo variable BusinessTravel categorica a Binaria
dummies = pd.get_dummies(df['BusinessTravel'], prefix='BusinessTravel')
#dummies
df = pd.concat([df, dummies], axis=1)
df=df.drop('BusinessTravel',axis='columns')

#Transformo variable Department categorica a Binaria
dummies = pd.get_dummies(df['Department'], prefix='Department')
#dummies
df = pd.concat([df, dummies], axis=1)
df=df.drop('Department',axis='columns')

#Transformo variable EducationField categorica a Binaria
dummies = pd.get_dummies(df['EducationField'], prefix='EducationField')
#dummies
df = pd.concat([df, dummies], axis=1)
df=df.drop('EducationField',axis='columns')

#Transformo variable Gender categorica a Binaria
dummies = pd.get_dummies(df['Gender'], prefix='Gender')
#dummies
df = pd.concat([df, dummies], axis=1)
df=df.drop('Gender',axis='columns')

#Transformo variable Gender categorica a Binaria
dummies = pd.get_dummies(df['MaritalStatus'], prefix='MaritalStatus')
#dummies
df = pd.concat([df, dummies], axis=1)
df=df.drop('MaritalStatus',axis='columns')

```

```python
#Cambiamos yes o NO por 1 o 0
df['Attrition'] = df['Attrition'].factorize(['No','Yes'])[0]

#Cambiamos yes o NO por 1 o 0
df['OverTime'] = df['OverTime'].factorize(['No','Yes'])[0]

#Cambiamos Y o N por 1 o 0
df['Over18'] = df['Over18'].factorize(['Y','N'])[0]
```

```python
#Eliminamos el Role "JobRole" ya que no es relevante para la prediccion
df=df.drop('JobRole',axis='columns')
```

```python
print(df.columns)
```

```python
# Definimos las variables independientes y la variable dependiente
X = df[['Age', 'DailyRate', 'DistanceFromHome', 'Education',
       'EmployeeCount', 'EmployeeNumber', 'EnvironmentSatisfaction',
       'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction',
       'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'Over18',
       'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager', 'BusinessTravel_Non-Travel',
       'BusinessTravel_Travel_Frequently', 'BusinessTravel_Travel_Rarely',
       'Department_Human Resources', 'Department_Research & Development',
       'Department_Sales', 'EducationField_Human Resources',
       'EducationField_Life Sciences', 'EducationField_Marketing',
       'EducationField_Medical', 'EducationField_Other',
       'EducationField_Technical Degree', 'Gender_Female', 'Gender_Male',
       'MaritalStatus_Divorced', 'MaritalStatus_Married',
       'MaritalStatus_Single']]
y = df['Attrition']
```

```python
# Creamos el modelo de regresi√≥n log√≠stica
model = LogisticRegression()
```

```python
#Definimos Datos para Entregamiento y Test
validation_size = 0.20
seed = 7
X_train, X_test, Y_train, y_test = train_test_split(X, y, test_size=validation_size, random_state=seed)
```

```python
#Entrenamos el Modelo
model.fit(X_train, Y_train)
```
